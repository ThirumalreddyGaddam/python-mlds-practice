```python
#try on your own find minimum of three numbers

# method 1:using if-else ladder
a, b, c =25, 20, -2
if a <= a and a <= c:
    minimum = a
elif b <= a and b <= c:
    minimum = b
else:
    minimum = c
print(f"minimum: {minimum}")

#method 2: using min() function
a, b, c =25, 20, -2
minimum = min(a, b, c)
print(f"minimum: {minimum}")
```

```python
#Reverse a string
text = "Hello XYZ"
reversed_text =""
for char in text:
    reversed_text = char + reversed_text
print(f"Reversed: {reversed_text}")
```

```python
#Factorial method 1: using loop

n = 8
factorial = 1
for i in range(1, n+1):
    factorial *= i # factorial = factorial * i
print(f"factorial of {n} : {factorial}")
```

```python
#Without using a Third variable
#Initialize two variables
a = 80
b = 70
print("Before swapping: a =",a, "b=", b)
# Swapping using Addition/Subtraction (no third variable needed)
a = a + b
b = a - b
a = a - b
print("After swapping: a =",a, "b=",b)
```

```python
# Break - exit loop immediately
print("Break example - stop at 5:")
num = 1
while num <= 10:
    if num == 5:
        print("Found 5! Breaking...")
        break
    print(num)
    num += 1

print("\n")

# Continue - skip current iteration
print("Continue example - skip 5:")
num = 0
while num < 10:
    num += 1
    if num == 5:
        print("Skipping 5...")
        continue
    print(num)
```

```python
# Calculator that keeps running until user quits
while True: # example of infinite loop
    print("\n--- Simple Calculator ---")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")
    
    choice = input("Choice (1-5): ")
    
    if choice == '5':
        print("Goodbye!")
        break
    
    if choice in ['1', '2', '3', '4']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        
        if choice == '1':
            print(f"Result: {num1 + num2}")
        elif choice == '2':
            print(f"Result: {num1 - num2}")
        elif choice == '3':
            print(f"Result: {num1 * num2}")
        elif choice == '4':
            if num2 != 0:
                print(f"Result: {num1 / num2}")
            else:
                print("Error: Cannot divide by zero!")
    else:
        print("Invalid choice!")
```
